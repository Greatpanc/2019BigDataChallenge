```python
import pandas as pd
path='/home/kesci/input/bytedance/bytedance_contest.final_2.csv'
print("正在读取：",path)
preds = pd.read_csv(path,
                    header=None,
                    usecols=[0,2],
                    names=['query_id','query_title_id']
                    )
print("长度为:",preds.__len__())

import numpy as np
path='/home/kesci/work/sunrui/nn/test2_pred/esim_pure_text_64_second1e_test2pred.npy'
print("正在读取：",path)
esim_pure_text_64_second1e_test2pred = np.load(path)
print("pres1 read done! len=",len(esim_pure_text_64_second1e_test2pred))

import numpy as np
path='/home/kesci/work/sunrui/nn/test2_pred/gated_esim_04_64_post1e_test2pred.npy'
print("正在读取：",path)
gated_esim_04_64_post1e_test2pred = np.load(path)
print("pres2 read done! len=",len(gated_esim_04_64_post1e_test2pred))

import numpy as np
path='/home/kesci/work/sunrui/nn/test2_pred/gated-01_esim_128_pre1e_test2pred.npy'
print("正在读取：",path)
gated_01_esim_128_pre1e_test2pred = np.load(path)
print("pres3 read done! len=",len(gated_01_esim_128_pre1e_test2pred))

import numpy as np
path='/home/kesci/work/sunrui/submit/gated-esim_128_epoch2_pre1e_test2pred.npy'
print("正在读取：",path)
gated_01_esim_128_pre1e_test2pred_new = np.load(path)
print("pres3 read done! len=",len(gated_01_esim_128_pre1e_test2pred_new))

import numpy as np
path='/home/kesci/work/sunrui/nn/test2_pred/gated-01_esim_64_pre1e_test2pred.npy'
print("正在读取：",path)
gated_01_esim_64_pre1e_test2pred=np.load(path)
print("pres4 read done! len=",len(gated_01_esim_64_pre1e_test2pred))

import numpy as np
path='/home/kesci/work/Submission/LGBRank_pre1e_807_final_baseline_3000_1_test2.npy'
print("正在读取：",path)
LGBRank_pre1e_807_final_baseline_3000_1_test2=np.load(path)
print("pres5 read done! len=",len(LGBRank_pre1e_807_final_baseline_3000_1_test2))

import numpy as np
def sigmoid(x):
    s = 1 / (1 + np.exp(-x))
    return s

preds["prediction"]=esim_pure_text_64_second1e_test2pred + \
                    gated_esim_04_64_post1e_test2pred + \
                    gated_01_esim_128_pre1e_test2pred + \
                    gated_01_esim_128_pre1e_test2pred_new + \
                    gated_01_esim_64_pre1e_test2pred + \
                    0.8*sigmoid(LGBRank_pre1e_807_final_baseline_3000_1_test2)

preds["prediction"]=preds["prediction"]/6
print("submission len =:",preds.__len__())
print("submission max =:",preds["prediction"].max())
print("submission min =:",preds["prediction"].min())
print("submission mean=:",preds["prediction"].mean())

SaveFile="/home/kesci/work/final_submission/LGBRank_sigmoid_w0.8_merge_5nn_final_test2.csv"
preds.to_csv(path_or_buf=SaveFile,header=False,index=0,encoding="utf-8")
print("save OK")
```
